Veri istatiksel model için dengeli mi

```python

def ikili_siniflandirici(etiket):
    # Eğer etiket 'normal' ise 0, değilse 1 döndürür
    if etiket == 'normal':
        return 0
    else:
        return 1
df['target'] = df['label'].apply(ikili_siniflandirici)

# Çalışıp çalışmadığını ilk 10 satırda test edelim
print(df[['label', 'target']].head(10))
```

```python

def siniflandirici(satir):
    #Etiket zaten normal değilse bu kesin bir saldırıdır
    if satir['label'] != 'normal':
        return 1
    
    # Eğer gönderilen veri (src_bytes) > 5000 ise VE süre (duration) < 1 saniye ise
    # Bu, normal bir kullanıcı için sıra dışı bir veri girişi  olabilir.
    if satir['src_bytes'] > 5000 and satir['duration'] < 1:
        return 1
    # Hiçbir şüpheli duruma takılmıyorsa güvenli kabul et
    return 0
df['target'] = df.apply(siniflandirici, axis=1)

# Algoritmanın yakaladığı "normal etiketli ama şüpheli" trafiği kontrol edelim
supheli_trafik = df[(df['label'] == 'normal') & (df['target'] == 1)]
print(f"Normal etiketli olup algoritmanın 'şüpheli' bulduğu satır sayısı: {len(supheli_trafik)}")
```

Kural tabanlı bir algoritma geliştirerek, normal etiketli trafiğin içindeki %1.24'lük şüpheli veri akışını tespit ettim

Böylece modelin nadir gerçekleşen ama kritik öneme sahip olan (U2R gibi) yetki yükseltme saldırılarını yakalama hassasiyeti (Recall) yükseltilmiştir.
 Değişken sayısı ve sınıf karmaşıklığı nedeniyle iterasyon sınırı 2000'e çıkarılarak modelin en düşük hata noktasına (Global Minimum) ulaşması sağlanmıştır.
